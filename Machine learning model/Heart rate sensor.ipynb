# ONE HOT ENCODINT

```python
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([("abcd", OneHotEncoder(), [1])], remainder = 'passthrough')
X = ct.fit_transform(X)
X.shape
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
classifier1=LogisticRegression()
classifier1.fit(X_train,y_train)
```

```python
y_pred1=classifier1.predict(X_test)
```

```python
from sklearn.metrics import  accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
x=range(1,26)
acc=[]
for i in x:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    acc.append(accuracy_score(y_test,y_pred))
plt.plot(x,acc)
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
y_pred2=knn.predict(X_test)
accuracy_score(y_test,y_pred2)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
y_pred3=tree.predict(X_test)
accuracy_score(y_test,y_pred3)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=50)
forest.fit(X_train,y_train)
y_pred4=forest.predict(X_test)
accuracy_score(y_test,y_pred4)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred4)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred4))
```

```python
def predict_input():
    age = int(input("Enter age "))
    sex = int(input("Enter Sex "))
    resting_bp_s = int(input("Enter resting blood sugar "))
    cholesterol = int(input("Enter cholesterol "))
    max_heart_rate = int(input("Enter heart rate "))
    
    my_pridiction = [[(1-sex),sex,age,resting_bp_s,cholesterol,max_heart_rate]]
    y_pred4=forest.predict(my_pridiction)
    return y_pred4[0]
```
